````python
num = int(input('–í–≤–µ–¥–∏—Ç–µ —Ü–µ–ª–æ–µ —á–∏—Å–ª–æ:'))
if not(num % 2):
    print('–ß–∏—Å–ª–æ —á—ë—Ç–Ω–æ–µ')
else:
    print('–ß–∏—Å–ª–æ –Ω–µ—á—ë—Ç–Ω–æ–µ')
````

**–ü—Ä–∏–º–µ—Ä 2**. –ù–∞–ø–∏—Å–∞—Ç—å —Å–∫—Ä–∏–ø—Ç, –∫–æ—Ç–æ—Ä—ã–π –∑–∞–ø—Ä–∞—à–∏–≤–∞–µ—Ç —É –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—è –≤–æ–∑—Ä–∞—Å—Ç, —Å—Ä–∞–≤–Ω–∏–≤–∞–µ—Ç —Å –ø–æ—Ä–æ–≥–æ–º –≤ –ø–µ—Ä–µ–º–µ–Ω–Ω–æ–π `threshold`. –ü—Ä–æ–≥—Ä–∞–º–º–∞ –¥–æ–ª–∂–Ω–∞ –ø—Ä–∏–≤–µ—Ç—Å—Ç–≤–æ–≤–∞—Ç—å –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—è –µ—Å–ª–∏ –µ–≥–æ –≤–æ–∑—Ä–∞—Å—Ç –Ω–µ –º–µ–Ω—å—à–µ –ø–æ—Ä–æ–≥–∞ –∏ –≤—ã–≤–æ–¥–∏—Ç—å —Å–æ–æ–±—â–µ–Ω–∏–µ "–î–æ—Å—Ç—É–ø –∑–∞–ø—Ä–µ—â–µ–Ω!" –≤ –ø—Ä–æ—Ç–∏–≤–Ω–æ–º —Å–ª—É—á–∞–µ.

````python
threshold = 18
age = int(input('–í–≤–µ–¥–∏—Ç–µ –≤–æ–∑—Ä–∞—Å—Ç:'))
if age >= threshold:
    print('–î–æ–±—Ä–æ –ø–æ–∂–∞–ª–æ–≤–∞—Ç—å!')
else:
    print('–î–æ—Å—Ç—É–ø –∑–∞–ø—Ä–µ—â–µ–Ω!')
````

**–ü—Ä–∏–º–µ—Ä 3**. –í—ã—è—Å–Ω–∏—Ç—å, –ø–æ–ø–∞–ª–∞ –ª–∏ —Ç–æ—á–∫–∞ $(x, y)$ –≤ –∫—Ä—É–≥ —Å —Ü–µ–Ω—Ç—Ä–æ–º –≤ –Ω–∞—á–∞–ª–µ –∫–æ–æ—Ä–¥–∏–Ω–∞—Ç –∏ —Ä–∞–¥–∏—É—Å–æ–º $R$. –°–º. —É—Ä–∞–≤–Ω–µ–Ω–∏–µ –æ–∫—Ä—É–∂–Ω–æ—Å—Ç–∏.

````python
x, y, R = float(input('x = ')), float(input('y = ')), float(input('R = '))
if (x * x + y * y)**0.5 <= R:
    print("–ü–æ–ø–∞–¥–∞–Ω–∏–µ")
else:
    print("–ü—Ä–æ–º–∞—Ö")
````

–†–µ—à–∏–º —ç—Ç—É –∑–∞–¥–∞—á—É –∏—Å–ø–æ–ª—å–∑—É—è **—Ç–µ—Ä–Ω–∞—Ä–Ω—ã–π –æ–ø–µ—Ä–∞—Ç–æ—Ä** Python:

```python
–∑–Ω–∞—á–µ–Ω–∏–µ_–µ—Å–ª–∏_–∏—Å—Ç–∏–Ω–∞ if —É—Å–ª–æ–≤–∏–µ else –∑–Ω–∞—á–µ–Ω–∏–µ_–µ—Å–ª–∏_–ª–æ–∂—å
```
–ü—Ä–∏–º–µ—Ä:
```python
result = a if condition else b
```

```python
if condition:
    result = a
else
    result = b
```

````python
x, y, R = float(input("x = ")), float(input("y = ")), float(input("R = "))
print('–ü–æ–ø–∞–¥–∞–Ω–∏–µ' if (x*x + y*y)**0.5 <= R else '–ü—Ä–æ–º–∞—Ö')
````

### –ü–µ—Ä–µ–¥–∞—á–∞ —É–ø—Ä–∞–≤–ª–µ–Ω–∏—è

**–ü—Ä–∏–º–µ—Ä 3**. –í—ã–≤–µ—Å—Ç–∏ –ø–æ—Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å–Ω–æ—Å—Ç—å –Ω–∞—Ç—É—Ä–∞–ª—å–Ω—ã—Ö –Ω–µ—á–µ—Ç–Ω—ã—Ö —á–∏—Å–µ–ª –Ω–µ –ø—Ä–µ–≤–æ—Å—Ö–æ–¥—è—â–∏—Ö 10.

````python
# –ü—Ä–∏–º–µ—Ä –¥–µ–º–æ–Ω—Å—Ç—Ä–∏—Ä—É–µ—Ç –∏—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω–∏–µ –æ–ø–µ—Ä–∞—Ç–æ—Ä–∞ continue, –∑–∞–¥–∞—á—É –º–æ–∂–Ω–æ —Ä–µ—à–∏—Ç—å —ç—Ñ—Ñ–µ–∫—Ç–∏–≤–Ω–µ–µ
i = 0
while i < 10:
    i += 1 #i = i + 1 
    if i % 2 == 0:
        continue
    print(i, end=' ')
    
````

–° –ø–æ–º–æ—â—å—é –æ–ø–µ—Ä–∞—Ç–æ—Ä–∞ while –º–æ–∂–Ω–æ –æ—Ä–≥–∞–Ω–∏–∑–æ–≤–∞—Ç—å –≤–≤–æ–¥ –ø–æ—Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å–Ω–æ—Å—Ç–∏, –∫–æ—Ç–æ—Ä–∞—è –æ–∫–∞–Ω—á–∏–≤–∞–µ—Ç—Å—è 0:

````python
# –≠—Ç–æ –±–µ—Å–ø–æ–ª–µ–∑–Ω—ã–π, –Ω–æ –∫–æ—Ä–æ—Ç–∫–∏–π –ø—Ä–∏–º–µ—Ä
while int(input()):
    print('–ù–∞–¥–æ–µ—Å—Ç, –≤–≤–µ–¥–∏—Ç–µ 0...')
````

````python
# –ú–æ–¥—É–ª—å random –ø—Ä–µ–¥–æ—Å—Ç–∞–≤–ª—è–µ—Ç —Ñ—É–Ω–∫—Ü–∏–∏ –¥–ª—è –≥–µ–Ω–µ—Ä–∞—Ü–∏–∏ —Å–ª—É—á–∞–π–Ω—ã—Ö —á–∏—Å–µ–ª –∏ –Ω–µ —Ç–æ–ª—å–∫–æ.
import random
# random.seed(3)

n = int(input('–í–≤–µ–¥–∏—Ç–µ n: '))
summ = 0
while summ < n:
    rnd_num = random.randint(0, 1) # –¢–µ–ø–µ—Ä—å –º–æ–∂–Ω–æ –∏—Å–ø–æ–ª—å–∑–æ–≤–∞—Ç—å –≤—Å–µ —Ñ—É–Ω–∫—Ü–∏–∏ –º–æ–¥—É–ª—è random
    summ += rnd_num
    print(rnd_num, end=' ')   
````

````python
def my_sum(iterable):
    '''–í—ã—á–∏—Å–ª—è–µ—Ç —Å—É–º–º—É –∏—Ç–µ—Ä–∏—Ä—É–µ–º–æ–≥–æ –∫–æ–Ω—Ç–µ–π–Ω–µ—Ä–∞'''
    summ = 0
    for elem in iterable:
        summ += elem
    return summ

def test_summ(func):
    '''–§—É–Ω–∫—Ü–∏—è —Ç–µ—Å—Ç–∏—Ä–æ–≤–∞–Ω–∏—è —Ñ—É–Ω–∫—Ü–∏–∏ my_sum'''
    print('test 1', 'ok' if func([1,2,5]) == 8 else 'fail')
    print('test 2', 'ok' if func([]) == 0 else 'fail')
    print('test 3', 'ok' if func({-3,4,-5,6,8}) == 10 else 'fail')
    print('test 4', 'ok' if func({-3:5, 4:'7'}) == 1 else 'fail')

if __name__ == '__main__':
    test_summ(my_sum)
````

**–ü—Ä–∏–º–µ—Ä 2**. –†–µ–∞–ª–∏–∑–æ–≤–∞—Ç—å –∞–Ω–∞–ª–æ–≥ —Ñ—É–Ω–∫—Ü–∏–∏ map, –≤–æ–∑–≤—Ä–∞—â–∞—é—â–∏–π —Å–ø–∏—Å–æ–∫.

````python
def map_func(function, iterable):
    '''–ü—Ä–∏–º–µ–Ω—è–µ—Ç —Ñ—É–Ω–∫—Ü–∏—é –∫ –∫–∞–∂–¥–æ–º—É —ç–ª–µ–º–µ–Ω—Ç—É –∫–æ–Ω—Ç–µ–π–Ω–µ—Ä–∞ –∏ –≤–æ–∑–≤—Ä–∞—â–∞–µ—Ç —Å–ø–∏—Å–æ–∫'''
    result = []
    for elem in iterable:
        result.append(function(elem))
    return result

map_func(int, [2.3, 4.7, -2, 4.6])
````

````python
def fib_rec(n):
    if n < 0:
        raise Exception("Negative value!")
    elif n == 0:
        return 0
    elif n <= 2:
        return 1
    else:
        return fib_rec(n-2) + fib_rec(n-1)

fib_rec(0), fib_rec(1), fib_rec(2), fib_rec(3), fib_rec(4), fib_rec(5), fib_rec(6),
````
