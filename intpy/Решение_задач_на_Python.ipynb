```python
def electronic_clock(num):
    minutes = num % 60
    hours = (num // 60) % 24
    return str(hours) + ' ' + str(minutes)
a = electronic_clock(1441)
print(a)
```

### Задача 2. Стрелочные часы

Написать код, который будет рассчитывать угол между часовой и минутной стрелкой в заданное время

```python
from datetime import datetime

# зачем нужны комментарии к функциям :)
def calc_angel(arg):
    """
        Рассчитывает угол между часовой и минутной стрелкой в заданное время;
        Input: arg - datetime, время;
        Output: число, угол
    """
    h = arg.hour
    m = arg.minute
    if h > 12:
        h = h - 12 # h -=12
    h_angle = 0.5 * (h * 60 + m)
    m_angle = 6 * m
    ans = abs(h_angle - m_angle)
    return min(ans, 360 - ans)

calc_angel(datetime.strptime('01:00', '%H:%M'))
```

### Задача 5. Игра "Угадай число"

Суть игры: мы загадываем число, а компьютер пытается его отгадать. Для решения предлагается использовать самый эффективный метод - бинарный поиск. Таким образом, компьютер сможет на каждой попытке как бы справшивать нас "больше названное число или меньше загаданного".

```python
def game_score(number):
    """
        Компьютер отгадывает число number (от 0 до 100);
        Input: число, которое мы загадали
        Output: число попыток, за комп отгадал число
    """
    counter = 1
    minimum = 0
    maximum = 100
    predict = (minimum + maximum) // 2
    while number != predict:
        counter += 1
        if number > predict:
            minimum = predict
        else:
            maximum = predict
        predict = (minimum + maximum) // 2
    return counter

game_score(76)
```
