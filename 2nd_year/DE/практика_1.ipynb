```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from selenium.common.exceptions import InvalidSessionIdException
from selenium.webdriver.common.by import By
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import TimeoutException
import time
from selenium.common.exceptions import NoSuchElementException
```

```python
# Сайт, с которым мы будем работать
url = "https://tproger.ru/quiz/real-programmer"

# функция для открытия сайта
def open_website(url):
    try:
        service = Service(ChromeDriverManager().install())
        driver = webdriver.Chrome(service=service)
        driver.get(url)
        return driver
    except InvalidSessionIdException:
        print("Session ID is invalid. Trying to restart the browser...")
        driver.quit()
        driver = webdriver.Chrome(service=service)
        driver.get(url)
        return driver
    
# функция для закрытия сайта
def close_browser(driver):
    driver.close()
    driver.quit()
```

```python
# Парсим вопросы  и правильные ответы к ним 

# Переходим на сайт
driver = open_website(url)
    
# Объявляем список для записи вопросов
questions = []

# Объявляем список для записи правильных ответов
right_answers = []

# индикатор наличия кнопки после окончания теста - тест не окончен/окончен -True/False
count = True

while(count):
    question = WebDriverWait(driver, 10).until(
    EC.presence_of_element_located((By.CLASS_NAME, "tp-quiz-question__text"))
    )
    questions.append(question.text)
    
    # нажимаем первую кнопку с ответом
    bot_1 = driver.find_element(By.CSS_SELECTOR, "input.tp-answer__radio")
    driver.execute_script("arguments[0].click()", bot_1)
    
    time.sleep(1)
    
    right_answer = WebDriverWait(driver, 10).until(
    EC.presence_of_element_located((By.CLASS_NAME, "tp-answer--result-true"))
    )
    right_answers.append(right_answer.text)

    # нажимаем далее
    bot_2 = driver.find_element(By.CSS_SELECTOR, "span.tp-button__text")
    driver.execute_script("arguments[0].click()", bot_2)
    
    time.sleep(1)
    
    # проверяем наличие на странице кнопки после окончания теста
    element = driver.find_elements(By.XPATH, '//*[@id="quiz-app"]/div/div/div[3]/div/form/div[3]/button/div/span')
    if element:
        count = False
        
# закрываем сайт
close_browser(driver)
```

```python
# Распечатываем результаты парсинга

for i in range(len(questions)):
    print('Вопрос теста: ',questions[i])
    print('Правильный ответ',right_answers[i])
    print('\n\n')
```
