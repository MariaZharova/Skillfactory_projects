# 2. Последовательности
## Длина наибольшей подпоследовательности
Даны две последовательности, требуется найти длину их наибольшей общей подпоследовательности.

```python
n = int(input('Введите длину первой последовательности: '))  # Длина первой последовательности
a = [int(x) for x in input('Введите члены первой последовательности: ').split()]  # Содержимое первой последовательности
m = int(input('Введите длину второй последовательности: '))  # Длина второй последовательности
b = [int(x) for x in input('Введите члены второй последовательности: ').split()]  # Содержимое второй последовательности

dp = [[0] * (m + 1) for c in range(n + 1)]  # Формирование матрицы нулей
for i in range(1, n + 1):  # Заполнение матрицы
    for j in range(1, m + 1):
        dp[i][j] = max(dp[i - 1][j], dp[i][j - 1])
        if a[i - 1] == b[j - 1]:  # Если
            dp[i][j] = max(dp[i][j], dp[i - 1][j - 1] + 1)
print(dp)  # Итоговая матрица
print(dp[n][m])  # Ответ
```

## Длина наибольшей возрастающей подпоследовательности

Дана последовательность, требуется найти длину ее наибольшей возрастающей подпоследовательности. Подпоследовательностью последовательности называется некоторый набор ее элементов, не обязательно стоящих подряд.

```python
n = int(input('Введите длину последовательности: '))  # Длина последовательности
a = [int(x) for x in input('Введите члены последовательности: ').split()]  # Содержимое последовательности
d = [1] * len(a)  # Список единиц размера последовательности
for i in range(1, len(a)):
    for j in range(i):
        if a[j] < a[i]:  # Если элемент a[j] < a[i]
            d[i] = max(d[i], 1 + d[j])  # d[i] как максимум из текущего и 1 + d[j]
print(*d)  # Итоговый список
print(max(d))  # Ответ
```

## Числа Фибоначчи

```python
def fib(n):  # Определяем функцию для вычисления числа Фибоначчи
    if n in (0, 1):  # Расчёт для первых двух чисел
        return 1
    else:
        return fib(n-1) + fib(n-2)  # Рекурсивный вызов для расчёта числа по формуле
```

```python
class Fib:  # Создаем класс для функции
    cache = {}  # Инициализируем кэш

    def fib(self, n):  # Определяем функцию для вычисления числа Фибоначчи
        if n not in self.cache:  # Если n не найден в кэше
            if n in (0, 1):
                res = 1	# Расчёт для первых двух чисел
            else:
                res = self.fib(n-1) + self.fib(n-2)  # Рекурсивный вызов для расчёта числа по формуле
            self.cache[n] = res  # Сохранение результата в кэш
        return self.cache[n]  # Если n найден в кэше, то вернуть его

    def __call__(self, n):  # Делаем класс функцией
        return self.fib(n)
```

## Задача о рюкзаках

```python
N, M = [int(x) for x in input('Введите N и M: ').split()]  # Инициализация ввода
m = [0] + [int(x) for x in input('Введите значения масс слитков: ').split()]

dp = [[False] * (M + 1) for _ in range(N + 1)]  # Формирование структуры "рюкзака" с заполнением False
dp[0][0] = True  # Инициализация первой ячейки как True
for i in range(1, N + 1):
    for j in range(0, M + 1):
        dp[i][j] = dp[i - 1][j]
        if m[i] <= j and dp[i - 1][j - m[i]]:
            dp[i][j] = True  # Присвоение элементу значения True

inds = []  # Инициализация массива индексов
for j in range(M + 1):
    if dp[N][j]:
        inds.append(j)
print(max(inds))  # Получение ответа
```

# 3. Одномерная динамика
## Последовательности из 0 и 1
Требуется подсчитать количество последовательностей длины N, состоящих из 0 и 1, в которых никакие две единицы не стоят рядом.

```python
n = int(input('Введите длину подпоследовательности: '))  # Ввод числа
a = [1, 2]  # Инициализация списка вариантов

for i in range(2, n + 1):
    a.append(a[i - 1] + a[i - 2])  # Добавление суммы двух предыдущих значений к списку
print(a[n])  # Ответ
```

## Про мальчика
Мальчик подошел к платной лестнице. Чтобы наступить на любую ступеньку, нужно заплатить указанную на ней сумму. Мальчик умеет перешагивать на следующую ступеньку, либо перепрыгивать через ступеньку. Требуется узнать, какая наименьшая сумма понадобится мальчику, чтобы добраться до верхней ступеньки.

```python
N = int(input('Введите количество ступеней: '))  # Количество ступеней
costs = [int(x) for x in input('Введите стоимость каждой ступеньки: ').split()]  # Цена ступеней
if N == 1:  # Если ступень одна
    print(costs[0])
else:  # Если не одна
    m = [0] * N  # Создаем массив нулей по числу ступеней
    m[0], m[1] = costs[0], costs[1]  # Присваиваем первые два значения
    for i in range(2, N):  # Для остальных ступеней
        m[i] = min(m[i - 1], m[i - 2]) + costs[i]  # Вычисляем величину

    print(m[-1])  # Ответ
```

## Про дощечки
В дощечке в один ряд вбиты гвоздики. Любые два гвоздика можно соединить ниткой. Требуется соединить некоторые пары гвоздиков ниточками так, чтобы к каждому гвоздику была привязана хотя бы одна ниточка, а суммарная длина всех ниточек была минимальна.

```python
n = int(input('Введите количество гвоздей: '))  # Количество гвоздей
v = sorted([int(x) for x in input('Введите координаты всех гвоздей: ').split()])  # Координаты гвоздей

a = [0 for _ in range(len(v))]  # Минимальное количество связей для того, чтобы связать i гвоздей
for i in range(1, n):
    if i > 2:  # Для нормального случая
        a[i] = min(a[i - 1], a[i - 2]) + (v[i] - v[i - 1])
    else:  # Для случая с двумя гвоздями
        a[i] = a[i - 1] + (v[i] - v[i - 1])

print(a[n - 1])  # Ответ
```

# 4. Двумерная динамика
## Про таблицу
В прямоугольной таблице NxM в начале игрок находится в левой верхней клетке. За один ход ему разрешается перемещаться в соседнюю клетку либо вправо, либо вниз (влево и вверх перемещаться запрещено). Посчитайте, сколько есть способов у игрока попасть в правую нижнюю клетку.

```python
N, M = [int(x) for x in input('Введите N, M: ').split()]  # Ввод размеров таблицы
A = [[1] * M for _ in range(N)]  # Формирование таблицы из единиц
for i in range(1, N):
    for j in range(1, M):
        A[i][j] = A[i - 1][j] + A[i][j - 1]  # Заполнение таблицы

print(A[-1][-1])  # Ответ
```

## Про треугольник Паскаля
Треугольник Паскаля строится следующим образом. Первая строка состоит из одного числа, равного единице. Каждая следующая содержит на одно число больше, чем предыдущая. Первое и последнее из этих чисел равны 1, а все остальные вычисляются как сумма числа, стоящего в предыдущей строке над ним и числа, стоящего в предыдущей же строке слева от него.

```python
N = int(input('Введите число строк: '))  # Ввод числа строк треугольника Паскаля
if N >= 1:
    pascal = [[1]]  # Инициализация структуры треугольника
    for i in range(N - 1):
        row = pascal[-1]  # Текущая строка - последняя в структуре треугольника
        newrow = [1]  # Инициализация новой строки
        for j in range(len(row) - 1):
            newrow.append(row[j] + row[j + 1])  # Добавление к строке вычисленного содержимого
        newrow.append(1)  # Добавление в конец строки единицы
        pascal.append(newrow)  # Добавление сгенерированной строки к массиву треугольника

    for x in pascal:  # Вывод ответа в цикле
        print(*x)
```

## Про килограммы еды
В каждой клетке прямоугольной таблицы N×M записано некоторое число. Изначально игрок находится в левой верхней клетке. За один ход ему разрешается перемещаться в соседнюю клетку либо вправо, либо вниз (влево и вверх перемещаться запрещено). При проходе через клетку с игрока берут столько килограммов еды, какое число записано в этой клетке (еду берут также за первую и последнюю клетки его пути).

```python
N, M = [int(x) for x in input('Введите N, M: ').split()]  # Ввод размеров таблицы
A = [[int(x) for x in input('Введите размеры штрафов: ').split()] for _ in range(N)]  # Ввод размеров штрафов
B = [[0] * M for _ in range(N)]  # Матрица нулей размером MxN для хранения ответа

B[0][0] = A[0][0]  # Присвоение значения первой ячейки
for i in range(1, N):  # Циклы для заполнения первой строки и первого столбца
    B[i][0] = B[i - 1][0] + A[i][0]
for j in range(1, M):
    B[0][j] = B[0][j - 1] + A[0][j]

for i in range(1, N):  # Заполнение остальных ячеек
    for j in range(1, M):
        B[i][j] = min(B[i][j - 1], B[i - 1][j]) + A[i][j]
print(B[-1][-1])  # ответ
```

## Про шахматную доску
На шахматной доске (8x8) стоит одна белая шашка. Сколькими способами она может пройти в дамки?

(Белая шашка ходит по диагонали, на одну клетку вверх-вправо или вверх-влево. Шашка проходит в дамки, если попадает на верхнюю горизонталь.)

```python
a = [[0 for j in range(8)] for i in range(8)]  # Генерация поля
j_s, i_s = [int(x) for x in input('Введите положение шашки: ').split()]  # Ввод положения шашки
a[i_s - 1][j_s - 1] = 1  # Установка шашки с учётом индексации с 0

for i in range(i_s, 8):  # Цикл для строк
    for j in range(8):  # Цикл для столбцов
        if i - 1 >= 0:
            if j - 1 >= 0 and j + 1 <= 7:
                a[i][j] = a[i - 1][j - 1] + a[i - 1][j + 1]
            elif j - 1 >= 0:
                a[i][j] = a[i - 1][j - 1]
            elif j + 1 <= 7:
                a[i][j] = a[i - 1][j + 1]

print(sum(a[-1]))  # Ответ
```
